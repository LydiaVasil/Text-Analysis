```python
def print_user_orders(user_name):
    user_data = ORDERS_BY_USER[user_name]
    if not user_data :
        print('No Order Data for {}'.format(user_name))
    for order_date in user_data:
        print(order_date)
        for order in user_data[order_date]:
            print("{} - {} - {} - {}".format(order['mfr_name'], order['mfr_part_number'], order['quantity'], order['item_name']))
        print()
print_user_orders('wbarker@ustaxcourt.gov')
```

```python
def get_list_of_orders(user_name):
    out = []
    user_orders = ORDERS_BY_USER[user_name]
    for order_date in user_orders:
        order_key = '{}_{}'.format(user_name, order_date)
        for order in user_orders[order_date]:
            order['order_id'] = order_key
            del(order['mfr_part_number_t'])
            del(order['is_os3'])
            out.append(order)
    return out
```

```python
def create_user_profile(user_name):
    temp_profile =  {
        'user_name': user_name,
        'total_number_of_orders':0,
        'total_amount_spent':0,
        'average_amount_spent_per_order':0,
        'average_item_count_per_order':0,
        'total_unique_items_bought':0,
        'total_items_bought':0,
        'bought_part_numbers':[],
    }
    order_temp = []
    user_orders = ORDERS_BY_USER[user_name]
    for order_date in user_orders:
        temp_profile['total_number_of_orders'] += 1
        temp_profile['total_amount_spent'] += sum([float(x['extended_price']) for x in user_orders[order_date]])
        items_in_order = [x['mfr_part_number'] for x in user_orders[order_date]]
        order_temp.append(items_in_order)
    temp_profile['average_amount_spent_per_order'] = temp_profile['total_amount_spent'] / temp_profile['total_number_of_orders']
    temp_profile['average_item_count_per_order'] = sum([len(x) for x in order_temp]) / len(order_temp)
    temp_profile['bought_part_numbers'] = [x for y in order_temp for x in y]
    temp_profile['total_items_bought'] = len(temp_profile['bought_part_numbers'])
    temp_profile['total_unique_items_bought'] = len(list(set(temp_profile['bought_part_numbers'])))
    return temp_profile
print(create_user_profile('wbarker@ustaxcourt.gov'))    
```
